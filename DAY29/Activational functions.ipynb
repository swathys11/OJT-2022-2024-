Build and visualise Activational Functions for Neural Network

. Sigmoid Function
. Hyperbolic Tangent(tanh) Function
. Rectified Linear Unit (ReLu)
. Leaky Rectified Linear Unit (Leaky ReLu) 
. Softmax Function

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#create the independent variable x to plot all thede functions
#Generate input values
x = np.linspace(-10, 10, 100)
```

```python
#Define step function
def stepFunction(z):
    return np.where(z >= 0, 1, 0)
#compute output
y_step = stepFunction(x)

#plot the graph
plt.plot(x, y_step, label='Step Function', color='b')
plt.title('Step Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

```python
#Define sigmoid function
def sigmoid(z):
    return 1/(1 + np.exp(-z))
#compute output
y_sigmoid = sigmoid(x)

#plot the graph
plt.plot(x, y_sigmoid, label='Sigmoid', color = 'b')
plt.title('Sigmoid')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

```python
#Define Hyperbolic  Tangent Function
def tanh(z):
    return np.tanh(z)
#Compute outputs
y_tanh = tanh(x)

#plot graph
plt.plot(x, y_tanh, label = 'Tanh', color = 'r')
plt.title('Tanh')
plt.xlabel('input')
plt.ylabel('output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Define ReLu Function
def relu(z):
    return np.maximum(0, z)
#compute output
y_relu = relu(x)
#plot the graph
plt.plot(x, y_relu, label='ReLu', color='g')
plt.title('ReLu')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Define leaky ReLu function
def leaky_relu(z, alpha=0.01):
    return np.where(z > 0, z, alpha * z)
#compute graph
y_leakyrelu = leaky_relu(x)

#plot graph
plt.plot(x, y_leakyrelu, label='Leaky ReLu', color='y')
plt.title('Leaky ReLU')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define Softmax Function
def softmax(z):
    exp_z = np.exp(z)
    return exp_z / np.sum(exp_z, axis=0)
# Compute outputs
y_softmax = softmax(x)

# Plot the graph
plt.plot(x, y_softmax, label='Softmax', color='m')
plt.title('Softmax')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```
